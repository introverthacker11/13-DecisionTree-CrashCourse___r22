## Decision Tree Machine Learning in Python
* Decision Tree is a Supervised Machine Learning Algorithm which is used for both Classification and Regression problems.
* Decision Tree is a tree-like structure which is used to represent the decisions and their possible consequences.
* It is flowchart-like structure in which each internel node donates a test on an attribute, each branch represent an outcome of the test, and each leaf node (terminal node) hold a class label.
* Decision Tree is a white box model, which means that if a given situation is observable in a model, the explanation for the condition is easily explained by boolean logic.

## The Process:

Root Node (Outlook):

The first decision is based on the Outlook feature, which has three possible values: sunny, overcast, or rainy.
If Outlook is:
* Sunny: The next decision is based on Humidity.
* Overcast: The decision is to play outside (Yes).
* Rainy: The next decision is based on whether it is Windy.

Second Layer of Decisions:

For the Sunny Outlook:
The tree splits based on Humidity:
If Humidity is High, the decision is No (dont play).
If Humidity is Normal, the decision is Yes (play).
For the Rainy Outlook:
The tree splits based on whether it is Windy:
If it is Windy (True), the decision is No (dont play).
If it is Not Windy (False), the decision is Yes (play).
